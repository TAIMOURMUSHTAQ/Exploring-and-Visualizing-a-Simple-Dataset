```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Loading the dataset 
iris = sns.load_dataset("iris")
# Overview
print("Shape:", iris.shape)
print("Columns:", iris.columns)
print("First 5 rows:", iris.head())
```

```python
# Info about dataset
print("Dataset Info:")
print(iris.info())
# Summary
print("Summary Stats:")
print(iris.describe())
```

```python
# Scatter plot: sepal_length vs petal length colored by species
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
```

```python
# Histogram for each feature
iris.hist(bins=15, figsize=(10,8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
# Boxplot for each feature grouped by species
plt.figure(figsize=(12,8))
sns.boxplot(data=iris)
plt.title("Boxplots of Iris Features")
plt.show()

# Species-wise boxplots
plt.figure(figsize=(12,6))
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Petal Length by Species")
plt.show()
```
